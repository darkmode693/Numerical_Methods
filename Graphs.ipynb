#Equations

```
## List of Polynomials Funcitons to be Tested
def a(x): return (x**3 + 5*x**2 + 13) ** 0.15 ## Used in Brute Force Method
def b1(x): return x**3 - 4*x**2 + 2*x + 3      ## Used in Newton Raphson Method
def b1_prime(x): return 3*x - 8*x + 2          ## Used in Newton Raphson Method
def c(x): return 2*x**2 - 5*x + 3             ## Used in Bisection Method 
def d(x): return x**3 + 4*x**2 - 3*x - 18     ## Used in Regula Falsi Method

## List of Trigonometric Functions to be Tested 
def e(x): return np.cos(x) + np.sin(x)        ## Used in Brute Force Method
def f(x): return np.sin(x) + np.cos(x)        ## Used in Newton Raphson Method
def f_prime(x): return np.cos(x) - np.sin(x)  ## Used in Newton Raphson Method
def g(x): return np.sin(x) - np.arccosh(x)    ## Used in Bisection Method
def h(x): return np.sin(x) - np.arccosh(x)    ## Used in Regula Falsi Method

## List of Logarithmic Functions to be Tested
def i(x): return np.log(x+2) - np.log(x+1)      ## Used in Brute Force Method
def j(x): return np.log(x + 2) + np.log(x - 1)  ## Used in Newton Raphson Method
def j_prime(x): return (2*x + 1)/(x-1)*(x+2)    ## Used in Newton Raphson Method
def k(x): return np.log(x-1) * (2*x - 5)        ## Used in Bisection Method  np.log(x**3-5*x-2)*(x**2-4)
def l(x): return np.log(x-1)*(x+5)              ## Used in Regula Falsi Method

## Final Equations
def m(x): 2*x**2 - 5*x + 3
```

#Bisection Method

### Using Polynomial Function

```
history_x = []
history_x_prime = []

# Single Root
i1, i2 = 1.1, 2
y1, y2 = b(i1), b(i2)
root = None
end_bisect = 0
if np.sign(y1) == np.sign(y2):
    print("Root cannot be found in the given interval")
else:
    for bisect in range(0,10):
        midp = np.mean([i1,i2])
        y_mid = b(midp)
        #y1 = f(i1)
        history_x.append(i1)
        history_x_prime.append(midp)
        if np.allclose(0,y1, 1e-06):
            root = i1
            end_bisect = bisect
            break
        if np.sign(y1) != np.sign(y_mid): #root is in first-half interval
            i2 = midp
        else: #root is in second-half interval
            i1 = midp
```

```
history_x = history_x[0:]
history_x_prime = history_x_prime[0:]
x_range = np.arange(len(history_x),dtype=int)

plt.figure(figsize=(10,7))
plt.plot(x_range, history_x, 'o--',color='dodgerblue', label='x')
plt.plot(x_range, history_x_prime,'D--', color='orange', label='x\'')

plt.title("Fig. 2.1 Convergent Solution")
plt.xlabel("Epochs")
plt.xticks(np.arange(min(x_range), max(x_range)+1, 1.0))
plt.legend()
plt.grid()
plt.show()
```
